```python
# Import our dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# replace with load from postgres
data_df = pd.read_csv('FINALCOMBOUSER1.csv')
data_df
```

```python
new_data_df = data_df.drop(["user_id", "latitude", "longitude", "placeid", "food_rating", "service_rating", "color", "weight", "height"], axis='columns')
new_data_df.head()
```

```python
n = new_data_df.nunique(axis=0)
print(n)
```

```python
cuisine = new_data_df['rcuisine']
cuisine.value_counts()
```

```python
clean_data_df = new_data_df.drop(["rcuisine"], axis='columns')
```

```python
cleaned_data_df = pd.get_dummies(clean_data_df)
cleaned_data_df.head()
```

```python
print(list(cleaned_data_df.columns))
```

```python
n = cleaned_data_df.nunique(axis=0)
print(n)
```

```python
y = cleaned_data_df["rating"].values
X = cleaned_data_df.drop("rating", axis=1)
```

```python
y
```

```python
X
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# scale data
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
clf = RandomForestClassifier(random_state=1, n_estimators=500, max_features='sqrt').fit(X_train_scaled, y_train)
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

```python
feature_importances = clf.feature_importances_
features = sorted(zip(X.columns, clf.feature_importances_), key = lambda x: x[1])
cols = [f[0] for f in features]
width = [f[1] for f in features]

fig, ax = plt.subplots()

fig.set_size_inches(10,200)
plt.margins(y=0.001)

ax.barh(y=cols, width=width)

plt.show()
```
