```python
# Import our dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# replace with load from postgres
data_df = pd.read_csv('FINALCOMBOUSER1.csv')
data_df
```

```python
new_data_df = data_df.drop(["user_id", "latitude", "longitude", "placeid", "food_rating", "service_rating"], axis='columns')
new_data_df.head()
```

```python
n = new_data_df.nunique(axis=0)
print(n)
```

```python
cuisine = new_data_df['rcuisine']
cuisine.value_counts()
```

```python
clean_data_df = new_data_df.drop(["rcuisine"], axis='columns')
```

```python
cleaned_data_df = pd.get_dummies(clean_data_df)
cleaned_data_df.head()
```

```python
print(list(cleaned_data_df.columns))
```

```python
n = cleaned_data_df.nunique(axis=0)
print(n)
```

```python
y = cleaned_data_df["rating"].values
X = cleaned_data_df.drop("rating", axis=1)
```

```python
y
```

```python
X
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# scale data
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
n_estimators = [20,50,100,200,300,500,700,900,1000] # number of trees in the random forest
max_features = ['auto', 'sqrt'] # number of features in consideration at every split
random_state = [1,2,4,8,10]


random_grid = {'n_estimators': n_estimators,
                'max_features': max_features,
                'random_state' : random_state}
```

```python
## Importing Random Forest Classifier from the sklearn.ensemble
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf_random = RandomizedSearchCV(estimator = rf,param_distributions = random_grid,
               n_iter = 90)
```

```python
# this is where we are testing the individual hyperparameters
rf_random.fit(X_train, y_train)
```

```python
print ('Random grid: ', random_grid, '\n')
# print the best parameters
print ('Best Parameters: ', rf_random.best_params_, ' \n')
```

```python
clf = RandomForestClassifier(random_state=4, n_estimators=20, max_features='auto').fit(X_train_scaled, y_train)
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
#Training Score: 0.8372379778051788
#Testing Score: 0.8027127003699137
```
