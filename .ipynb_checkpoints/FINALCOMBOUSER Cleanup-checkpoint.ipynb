```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# Read csv
combo_user = pd.read_csv('FINALCOMBOUSER1.csv')

# Convert csv to data frame
combo_user = pd.DataFrame(combo_user)

combo_user.head()
```

```python
combo_user.columns
```

```python
combo_user = combo_user.drop(columns={'color','weight','height','religion','interest','personality' })
```

```python
combo_user.columns
```

```python
combo_user = combo_user.rename(columns={'latitude':'Latitude',
                                        'longitude':'Longitude',
                                        'smoker':'Smoker',
                                        'drink_level':'Alcohol Consumption',
                                       'dress_perference':'Dress Code',
                                       'transport':'Transport',
                                       'marital_status':'Marital Status',
                                       'birth_year':'Birth Year',
                                       'activity':'Occupation',
                                       'budget':'Budget',
                                       'placeid':'Place ID',
                                       'user_id':'User ID',
                                       'rating':'Overall Rating',
                                       'food_rating':'Food Rating',
                                       'service_rating':'Service Rating',
                                       'rcuisine':'Cuisine',
                                       'ambience': 'Ambiance Preference',
                                       'hijos':'Children'})
```

```python
combo_user.columns
```

```python
combo_user.nunique()
```

```python
combo_user = combo_user.replace('abstemious','Moderate')
```

```python
combo_user_loc = combo_user.loc[(combo_user["User ID"] == '?')&(combo_user["Latitude"] == '?')&(combo_user["Longitude"] == '?')
                                &(combo_user["Smoker"] == '?')&(combo_user["Alcohol Consumption"] == '?')&(combo_user["Dress Code"] == '?')
                                &(combo_user["Ambiance Preference"] == '?')&(combo_user["Transport"] == '?')&(combo_user["Marital Status"] == '?')
                                &(combo_user["Children"] == '?')&(combo_user["Birth Year"] == '?')&(combo_user["Occupation"] == '?')
                                &(combo_user["Budget"] == '?')&(combo_user["Place ID"] == '?')&(combo_user["Overall Rating"] == '?')
                                &(combo_user["Food Rating"] == '?')&(combo_user["Service Rating"] == '?')&(combo_user["Cuisine"] == '?')]
combo_user_loc
                                
```

```python
combo_user = combo_user.loc[(combo_user["User ID"]!= '?')&(combo_user["Latitude"] != '?')&(combo_user["Longitude"] != '?')
                                &(combo_user["Smoker"] != '?')&(combo_user["Alcohol Consumption"] != '?')&(combo_user["Dress Code"] != '?')
                                &(combo_user["Ambiance Preference"] != '?')&(combo_user["Transport"] != '?')&(combo_user["Marital Status"] != '?')
                                &(combo_user["Children"] != '?')&(combo_user["Birth Year"] != '?')&(combo_user["Occupation"] != '?')
                                &(combo_user["Budget"] != '?')&(combo_user["Place ID"] != '?')&(combo_user["Overall Rating"] != '?')
                                &(combo_user["Food Rating"] != '?')&(combo_user["Service Rating"] != '?')&(combo_user["Cuisine"] != '?')]
combo_user
```
